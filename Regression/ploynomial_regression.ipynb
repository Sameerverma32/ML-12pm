```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer
```

```python
temperature = [[20],[22],[24],[26],[28],[30],[32]]
sales = [10,15,25,40,60,90,100]
```

```python
x = temperature
y = sales
```

```python
y
```

```python
plt.plot(temperature, sales, marker='o', linestyle='-', color='blue')
plt.title("Temperature vs Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    'Temperature': temperature,
    'Sales': sales
})
```

```python
# Train the dataset
lr = LinearRegression()
lr.fit(x, y)

```

```python
#Polynomial Regression
pr = PolynomialFeatures(degree=5)
X_poly = pr.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(X_poly, y)
```

```python
# predict
y_pred_poly = lr2.predict(X_poly)
print(y_pred_poly)

y_pred_lr = lr.predict(x)
print(y_pred_lr)
```

```python
# Visualize the linear regression
plt.scatter(x, y, color='red')
plt.plot(x, y_pred_lr, color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['X/y_pred_lr', 'X/y'], title="Salary/Level", loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Polynomial Regression
plt.scatter(x, y, color='red')
plt.plot(x, y_pred_poly, color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend(['X/y_pred_poly', 'X/y'], title="Salary/Level", loc='best', facecolor='white')
plt.box(False)
plt.show()
```
