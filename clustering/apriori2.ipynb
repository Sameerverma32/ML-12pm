```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('Groceries_dataset.csv')
df
```

```python
df_grouped = df.groupby(['Member_number','Date'])['itemDescription'].apply(list).reset_index()
df_grouped
```

```python
transaction = df_grouped['itemDescription'].tolist()
```

```python
te = TransactionEncoder()
df_encoded = te.fit(transaction).transform(transaction)
df_encoded = pd.DataFrame(df_encoded, columns=te.columns_)
df_encoded
```

Support tells us how frequently an itemset appers in the dataset

```python
frequent_itemsets = apriori(df_encoded, min_support=0.002, use_colnames=True)
print
```
